```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os 
sns.set()
%matplotlib inline
warnings.filterwarnings("ignore")
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
combine = pd.concat([train, test])
```

```python
print(train.shape, test.shape)
print(train.info())
print(test.info())
print(train.isnull().sum(), test.isnull().sum())
```

```python
train.head()
```

```python
drop_id = ["id","amenities","description","first_review","host_has_profile_pic","host_response_rate"
           ,"host_since","last_review","latitude","longitude","name","zipcode"]
train = train.drop(drop_id, axis=1)
```

```python
train = train.dropna(subset=["beds", "bathrooms", "bedrooms", "host_identity_verified"])
train = train.drop("neighbourhood", axis=1)
```

```python
print(train["review_scores_rating"].describe())
print(train["review_scores_rating"].unique())
print(train["review_scores_rating"].value_counts())
```

```python
train["review_scores_rating"] = train["review_scores_rating"].fillna(train["review_scores_rating"].median())
```

```python
train.head()
```

```python
def convert_one_hot_TF(x):
    if x == "t":
        return 1
    elif x == "f":
        return 0
    
train["cleaning_fee"] = train["cleaning_fee"].apply(convert_one_hot_TF)
train["host_identity_verified"] = train["host_identity_verified"].apply(convert_one_hot_TF)
train["instant_bookable"] = train["instant_bookable"].apply(convert_one_hot_TF)
```

```python
def CheckNull(x):
    if x != "":
        return 1
    else:
        return 0
train["thumbnail_url"] = train["thumbnail_url"].apply(CheckNull)
```

```python
select_columns = ["bed_type", "cancellation_policy", "city", "property_type", "room_type"]
dummy_df = pd.get_dummies(train[select_columns], drop_first=True)

train = pd.concat([train, dummy_df], axis=1)
train = train.drop(select_columns, axis=1)
```

```python
train.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
y = train["y"]
X = train.drop("y", axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mse_test = mean_squared_error(y_test, y_pred)
rmse_test = np.sqrt(mse_test)
print(rmse_test)
```

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```
