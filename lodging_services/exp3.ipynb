```python
def plot_barplot_with_value_counts(data, column, target):
    print(data[column].value_counts())
    plt.figure(figsize=(16, 16))
    sns.barplot(x=data[column], y=data[target])
    plt.xticks(rotation=90)
    plt.show()
```

```python
plot_barplot_with_value_counts(row, "city", "y")
```

```python
plot_barplot_with_value_counts(row, "property_type", "y")
```
